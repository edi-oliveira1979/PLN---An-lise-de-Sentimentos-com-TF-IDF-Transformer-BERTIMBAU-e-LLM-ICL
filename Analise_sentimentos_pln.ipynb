# Projeto de PLN — Análise de Sentimentos em Reviews de Produtos

Este projeto investiga a tarefa de **classificação de sentimentos em português** a partir de avaliações reais de consumidores, utilizando uma abordagem comparativa entre modelos clássicos, embeddings e transformers.

O pipeline foi desenvolvido em ambiente **Google Colab**, com persistência em **Google Drive**, e estruturado para publicação em **GitHub como portfólio técnico**.

**Tecnologias principais:** Python, Scikit-learn, Hugging Face Transformers, SentenceTransformers, BERTimbau, Gemini API.

O trabalho posiciona-se como um estudo aplicado de PLN, com foco em **avaliação empírica de diferentes paradigmas de modelagem** em dados reais.

## Objetivo

O objetivo do projeto é construir e avaliar modelos capazes de classificar automaticamente o sentimento de avaliações textuais em três classes: **positivo, neutro e negativo**.

Busca-se responder:

- Qual abordagem apresenta melhor desempenho para o problema?
- Qual o impacto da representação textual (bag-of-words vs embeddings vs transformers)?
- Quais limitações surgem em cenários reais, especialmente com classes desbalanceadas?

O escopo contempla desde modelos clássicos até abordagens baseadas em deep learning e LLMs, com foco em **comparabilidade e rigor experimental**.

## Base de Dados

Foi utilizado o dataset **B2W-Reviews01**, contendo avaliações de produtos em português.

- Volume total após filtragem: **128.299 registros**
- Campos relevantes: texto da avaliação e nota (rating)
- Derivação de labels:
  - positivo
  - neutro
  - negativo

**Distribuição das classes:**

- Positivo: ~61%
- Negativo: ~26%
- Neutro: ~12%

Observa-se um **desbalanceamento significativo**, especialmente na classe neutra, o que influencia diretamente a avaliação dos modelos.

O dataset apresenta linguagem informal, variação lexical e ruído textual típico de dados reais.

## Metodologia

### Pré-processamento

- Normalização textual (lowercase)
- Remoção de ruído básico
- Criação de variável `text_clean`

### Split

- Estratégia: **estratificada**
- Treino: 70% | Validação: 10% | Teste: 20%

### Modelos avaliados

1. **TF-IDF + SVM**
2. **Embeddings (SentenceTransformer) + SVM**
3. **BERTimbau fine-tuned**
4. **LLM (Gemini — In-Context Learning)**

### Extensões

- Avaliação de **hard cases**
- Análise de consistência entre modelos
- Testes com variação de **temperatura em LLMs**

### Métricas

- **F1 macro (principal)**
- Accuracy

A F1 macro foi priorizada por penalizar desempenho desigual entre classes, especialmente relevante em dados desbalanceados.

## Resultados

### Desempenho geral

- **TF-IDF + SVM:** baseline forte e estável
- **Embeddings + SVM:** leve ganho semântico, porém inconsistência global
- **BERT fine-tuned:** melhor desempenho geral
- **LLM (Gemini):** competitivo em alguns cenários, porém menos consistente

### Métricas principais

- BERT apresentou maior **F1 macro (~0.75)** e melhor equilíbrio entre classes
- TF-IDF manteve alta **accuracy (~0.83)**, mas com viés para classes majoritárias

### Análise por classe

- Positivo: alto desempenho em todos os modelos
- Negativo: desempenho consistente
- Neutro: principal gargalo

### Hard cases

- Alto nível de divergência entre modelos
- Tendência geral de **polarização indevida**
- Casos neutros frequentemente classificados como positivos ou negativos

Esses resultados indicam que a dificuldade não está apenas no modelo, mas na própria natureza dos dados.

## Análise de Robustez

### Hard Cases

A análise de aproximadamente 2000 casos difíceis revelou:

- Baixa concordância entre modelos
- Alta ambiguidade semântica
- Presença de sinais contraditórios no texto

Modelos baseados em BERT e embeddings ainda apresentam limitações na distinção fina entre neutralidade e polaridade leve.

### Consistência entre modelos

- TF-IDF e BERT convergem em casos claros
- Divergência aumenta significativamente em casos neutros
- LLM apresenta maior variabilidade de saída

### Sensibilidade à temperatura (LLM)

- Temperaturas baixas (0.0–0.2): maior consistência
- Temperaturas mais altas: aumento de variabilidade sem ganho claro de qualidade

Isso evidencia que LLMs são sensíveis a parâmetros de inferência, impactando diretamente a confiabilidade.

## Discussão

### Modelos clássicos vs transformers

- TF-IDF permanece competitivo como baseline
- BERT supera consistentemente em cenários complexos
- Embeddings isolados não garantem ganho significativo

### Impacto do desbalanceamento

- Classe neutra sofre com baixa representatividade
- Modelos tendem a maximizar acerto em classes majoritárias
- F1 macro evidencia essa limitação estrutural

### Classe neutra como desafio central

- Baixa separabilidade semântica
- Presença de ambiguidade contextual
- Forte sobreposição com outras classes

### Comportamento dos LLMs

- Capacidade de generalização sem treino
- Sensibilidade a parâmetros (temperatura)
- Baixa previsibilidade em cenários ambíguos

### Convergência do BERT

- Melhora incremental com mais épocas
- Estabilidade maior em comparação com outros modelos
- Melhor captura de contexto linguístico

O conjunto de resultados reforça que desempenho médio não captura completamente a robustez do modelo.

## Conclusão

Os experimentos demonstram que:

- **BERT fine-tuned** é a abordagem mais robusta e consistente
- Modelos clássicos continuam relevantes como baseline eficiente
- A classe neutra permanece como principal desafio do problema

A análise de hard cases revelou que:

- O erro não é apenas de modelagem, mas também estrutural do dataset
- Ambiguidade semântica limita o desempenho máximo alcançável

**Limitações:**

- Avaliação limitada de LLMs em larga escala

O projeto evolui de comparação de modelos para uma análise mais profunda de **robustez e comportamento em cenários reais**.

````python
# 1.4 Imports e utilitários

import re
import shutil
import pickle
import textwrap

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    classification_report,
    confusion_matrix,
)
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.manifold import TSNE
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE

plt.style.use("default")
sns.set_theme(style="whitegrid")

def save_json(obj, path: Path):
    def conv(x):
        if isinstance(x, (np.integer,)):
            return int(x)
        if isinstance(x, (np.floating,)):
            return float(x)
        if isinstance(x, np.ndarray):
            return x.tolist()
        raise TypeError(f"Tipo não serializável: {type(x)}")
    path.parent.mkdir(parents=True, exist_ok=True)
    with open(path, "w", encoding="utf-8") as f:
        json.dump(obj, f, ensure_ascii=False, indent=2, default=conv)

def load_json(path: Path):
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)

def save_fig(fig, filename: str):
    path = DIRS["outputs_figures"] / filename
    fig.savefig(path, dpi=160, bbox_inches="tight")
    print(f"💾 Figura salva em {path}")

def exists(path: Path) -> bool:
    return path.exists()

def rating_to_label(rating: int) -> str:
    if rating <= 2:
        return "negativo"
    if rating == 3:
        return "neutro"
    return "positivo"

def normalize_text(text: str) -> str:
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+", " ", text)
    text = re.sub(r"[^\w\sÀ-ÿ]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

def summarize_metrics(model_name: str, y_true, y_pred, extra=None):
    metrics = {
        "model": model_name,
        "f1_macro_test": float(f1_score(y_true, y_pred, average="macro")),
        "accuracy_test": float(accuracy_score(y_true, y_pred)),
        "classification_report": classification_report(
            y_true, y_pred, target_names=LABEL_NAMES, output_dict=True
        ),
    }
    if extra:
        metrics.update(extra)
    return metrics

print("✅ Utilitários prontos.")
````

## 2. Carga do dataset e entendimento inicial

Nesta versão,  foi utilizado o **B2W-Reviews01**, adequado ao escopo da disciplina:

````python
# 2.1 Download e cache do dataset

RAW_PATH = DIRS["data_raw"] / "b2w_reviews_raw.parquet"

def load_b2w_dataset_with_fallback(dataset_name: str) -> pd.DataFrame:
    """Carrega o dataset da HF Hub com fallback para os arquivos Parquet auto-convertidos."""
    from datasets import load_dataset
    import requests

    try:
        print("⬇️ Tentando carregar via load_dataset(...) ...")
        ds = load_dataset(dataset_name, split="train")
        return ds.to_pandas()

    except RuntimeError as e:
        msg = str(e)
        if "Dataset scripts are no longer supported" not in msg:
            raise

        print("⚠️ O dataset usa script legado no Hub. Aplicando fallback para Parquet...")

        api_url = f"https://datasets-server.huggingface.co/parquet?dataset={dataset_name}"
        resp = requests.get(api_url, timeout=60)
        resp.raise_for_status()
        payload = resp.json()

        parquet_files = payload.get("parquet_files", [])
        train_urls = [
            item["url"]
            for item in parquet_files
            if item.get("split") == "train"
        ]

        if not train_urls:
            raise RuntimeError(
                "Não foi possível localizar arquivos Parquet do split 'train' no endpoint /parquet."
            )

        print(f"📦 {len(train_urls)} arquivo(s) Parquet encontrado(s). Carregando...")
        parts = [pd.read_parquet(url) for url in train_urls]
        df = pd.concat(parts, ignore_index=True)
        return df

if exists(RAW_PATH):
    print("✅ Dataset encontrado em cache no Drive.")
    df_raw = pd.read_parquet(RAW_PATH)
else:
    print("⬇️ Baixando dataset do Hugging Face...")
    df_raw = load_b2w_dataset_with_fallback(CFG["dataset_hf"])
    df_raw.to_parquet(RAW_PATH, index=False)
    print(f"💾 Dataset salvo em {RAW_PATH}")

print(df_raw.shape)
display(df_raw.head(3))
````

````python
# 2.2 Preparação inicial e mapeamento de rótulos

df = df_raw[["review_text", "overall_rating"]].copy()
df = df.dropna(subset=["review_text", "overall_rating"]).reset_index(drop=True)
df["overall_rating"] = df["overall_rating"].astype(int)
df["text"] = df["review_text"].astype(str)
df["label"] = df["overall_rating"].apply(rating_to_label)
df["label_id"] = df["label"].map(LABEL2ID)

# Filtro simples de ruído
df = df[df["text"].str.split().str.len() >= CFG["min_words"]].reset_index(drop=True)

print(f"Total após filtro mínimo: {len(df):,}")
print(df["label"].value_counts())
display(df[["text", "overall_rating", "label"]].head(5))
````

````python
# 2.3 Pré-processamento textual

df["text_clean"] = df["text"].apply(normalize_text)
df = df[df["text_clean"].str.len() > 0].reset_index(drop=True)

df["n_words"] = df["text_clean"].str.split().str.len()
df["n_chars"] = df["text_clean"].str.len()

print("✅ Pré-processamento concluído.")
display(df[["text", "text_clean", "label"]].head(3))
````

````python
# 2.4 EDA rápida

if CFG["run_eda"]:
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))

    order = ["negativo", "neutro", "positivo"]
    sns.countplot(data=df, x="label", order=order, palette=PALETTE, ax=axes[0])
    axes[0].set_title("Distribuição de classes")

    sns.boxplot(data=df, x="label", y="n_words", order=order, palette=PALETTE, ax=axes[1])
    axes[1].set_title("Comprimento das avaliações (nº de palavras)")

    plt.tight_layout()
    save_fig(fig, "01_eda_distribuicao_e_tamanho.png")
    plt.show()
else:
    print("EDA desativada pela configuração.")
````

````python
# 3.1 Gerar ou carregar splits

TRAIN_PATH = DIRS["data_processed"] / "train.parquet"
VAL_PATH   = DIRS["data_processed"] / "val.parquet"
TEST_PATH  = DIRS["data_processed"] / "test.parquet"

if all(exists(p) for p in [TRAIN_PATH, VAL_PATH, TEST_PATH]):
    print("✅ Splits encontrados no Drive.")
    df_train = pd.read_parquet(TRAIN_PATH)
    df_val   = pd.read_parquet(VAL_PATH)
    df_test  = pd.read_parquet(TEST_PATH)
else:
    df_train_temp, df_test = train_test_split(
        df,
        test_size=CFG["test_size"],
        stratify=df["label_id"],
        random_state=SEED,
    )

    df_train, df_val = train_test_split(
        df_train_temp,
        test_size=CFG["val_size_over_train"],
        stratify=df_train_temp["label_id"],
        random_state=SEED,
    )

    df_train.to_parquet(TRAIN_PATH, index=False)
    df_val.to_parquet(VAL_PATH, index=False)
    df_test.to_parquet(TEST_PATH, index=False)

    print("💾 Splits salvos no Drive.")

for name, part in [("train", df_train), ("val", df_val), ("test", df_test)]:
    print(f"{name:>5}: {len(part):,}")
    print(part["label"].value_counts(normalize=True).sort_index())

X_train = df_train["text_clean"].tolist()
X_val   = df_val["text_clean"].tolist()
X_test  = df_test["text_clean"].tolist()

y_train = df_train["label_id"].to_numpy()
y_val   = df_val["label_id"].to_numpy()
y_test  = df_test["label_id"].to_numpy()
````

## 4. Modelo clássico — TF-IDF + LinearSVC

Esta é a referência mínima obrigatória do projeto.  
Ela costuma entregar um baseline forte, rápido e muito interpretável.

````python
# 4.1 Baseline TF-IDF + LinearSVC

TFIDF_BASE_PIPE_PATH = DIRS["models_tfidf"] / "tfidf_baseline.pkl"
TFIDF_BASE_METRICS_PATH = DIRS["outputs_metrics"] / "tfidf_baseline_metrics.json"

if CFG["run_tfidf"]:
    if exists(TFIDF_BASE_PIPE_PATH) and exists(TFIDF_BASE_METRICS_PATH):
        with open(TFIDF_BASE_PIPE_PATH, "rb") as f:
            tfidf_baseline = pickle.load(f)
        tfidf_baseline_metrics = load_json(TFIDF_BASE_METRICS_PATH)
        print("✅ Baseline TF-IDF carregado do cache.")
    else:
        tfidf_baseline = Pipeline([
            ("tfidf", TfidfVectorizer(
                lowercase=False,
                min_df=3,
                max_features=25000,
                ngram_range=(1, 2),
                sublinear_tf=True,
            )),
            ("clf", LinearSVC(C=1.0, class_weight="balanced", random_state=SEED)),
        ])
        tfidf_baseline.fit(X_train, y_train)
        y_pred_base = tfidf_baseline.predict(X_test)
        tfidf_baseline_metrics = summarize_metrics(
            "TF-IDF + LinearSVC (baseline)", y_test, y_pred_base
        )
        with open(TFIDF_BASE_PIPE_PATH, "wb") as f:
            pickle.dump(tfidf_baseline, f)
        save_json(tfidf_baseline_metrics, TFIDF_BASE_METRICS_PATH)
        print("💾 Baseline TF-IDF salvo.")

    y_pred_tfidf_base = tfidf_baseline.predict(X_test)
    print(json.dumps(
        {
            "f1_macro_test": round(tfidf_baseline_metrics["f1_macro_test"], 4),
            "accuracy_test": round(tfidf_baseline_metrics["accuracy_test"], 4),
        },
        ensure_ascii=False,
        indent=2,
    ))
else:
    print("Treino TF-IDF desativado.")
````

````python
# 4.2 GridSearch do modelo clássico

GRID_PIPE_PATH = DIRS["models_tfidf"] / "tfidf_grid_best.pkl"
GRID_METRICS_PATH = DIRS["outputs_metrics"] / "tfidf_grid_metrics.json"

if CFG["run_tfidf"]:
    if exists(GRID_PIPE_PATH) and exists(GRID_METRICS_PATH):
        with open(GRID_PIPE_PATH, "rb") as f:
            tfidf_best = pickle.load(f)
        metrics_tfidf = load_json(GRID_METRICS_PATH)
        print("✅ Melhor modelo TF-IDF carregado do cache.")
    else:
        grid_pipe = Pipeline([
            ("tfidf", TfidfVectorizer(lowercase=False, min_df=3, sublinear_tf=True)),
            ("clf", LinearSVC(class_weight="balanced", random_state=SEED)),
        ])

        param_grid = {
            "tfidf__max_features": [20000, 30000],
            "tfidf__ngram_range": [(1, 1), (1, 2)],
            "clf__C": [0.5, 1.0, 2.0],
        }

        gs = GridSearchCV(
            estimator=grid_pipe,
            param_grid=param_grid,
            scoring="f1_macro",
            cv=5,
            n_jobs=-1,
            verbose=1,
            refit=True,
        )
        gs.fit(X_train, y_train)

        best = gs.best_params_
        print("Melhores parâmetros:", best)

        tfidf_best = Pipeline([
            ("tfidf", TfidfVectorizer(
                lowercase=False,
                min_df=3,
                sublinear_tf=True,
                max_features=best["tfidf__max_features"],
                ngram_range=best["tfidf__ngram_range"],
            )),
            ("clf", LinearSVC(
                C=best["clf__C"],
                class_weight="balanced",
                random_state=SEED,
            )),
        ])

        tfidf_best.fit(X_train, y_train)
        y_pred_tfidf_test = tfidf_best.predict(X_test)
        metrics_tfidf = summarize_metrics(
            "TF-IDF + LinearSVC (grid)", y_test, y_pred_tfidf_test,
            extra={"best_params": best, "cv_best_score": float(gs.best_score_)},
        )

        with open(GRID_PIPE_PATH, "wb") as f:
            pickle.dump(tfidf_best, f)
        save_json(metrics_tfidf, GRID_METRICS_PATH)
        print("💾 Melhor modelo TF-IDF salvo.")

    y_pred_tfidf_test = tfidf_best.predict(X_test)
    print(classification_report(y_test, y_pred_tfidf_test, target_names=LABEL_NAMES, digits=4))
````

## 6. Transformer — BERTimbau fine-tuned

Este bloco é o mais caro computacionalmente.  
Por isso, foi deixado com **cache**, **controle por flag** e comportamento mais robusto no Colab.

````python
# 6.1 Preparação para fine-tuning BERT

import torch
from datasets import Dataset as HFDataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
)
import evaluate

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Dispositivo:", device)
if device == "cuda":
    print("GPU:", torch.cuda.get_device_name(0))
else:
    print("⚠️ GPU não detectada. Você pode manter run_bert=True e carregar cache, mas treinar do zero ficará lento.")

if CFG["run_bert"]:
    tokenizer = AutoTokenizer.from_pretrained(CFG["bert_model"])

    def tokenize_batch(batch):
        return tokenizer(
            batch["text"],
            truncation=True,
            padding="max_length",
            max_length=CFG["bert_max_len"],
        )

    def df_to_hf(df_part):
        ds = HFDataset.from_dict({
            "text": df_part["text_clean"].tolist(),
            "label": df_part["label_id"].tolist(),
        })
        ds = ds.map(tokenize_batch, batched=True, batch_size=256)
        ds.set_format("torch", columns=["input_ids", "attention_mask", "label"])
        return ds

    hf_train = df_to_hf(df_train)
    hf_val   = df_to_hf(df_val)
    hf_test  = df_to_hf(df_test)

    metric_f1 = evaluate.load("f1")
    metric_acc = evaluate.load("accuracy")

    def compute_metrics(eval_pred):
        logits, labels = eval_pred
        preds = np.argmax(logits, axis=-1)
        return {
            "f1_macro": metric_f1.compute(predictions=preds, references=labels, average="macro")["f1"],
            "accuracy": metric_acc.compute(predictions=preds, references=labels)["accuracy"],
        }

    print("✅ Datasets do Hugging Face prontos.")
````

````python
# 6.2 Continuação segura do BERT a partir do modelo final salvo

import inspect
import json
import numpy as np
from pathlib import Path

BERT_BASE_FINAL_DIR = DIRS["models_bert"] / "final"

# nova execução separada
RUN_TAG = CFG.get("bert_run_tag", "cont")
BERT_CONT_DIR = DIRS["models_bert"] / f"final_{RUN_TAG}"
BERT_CONT_CKPT_DIR = DIRS["models_bert"] / f"checkpoints_{RUN_TAG}"

BERT_METRICS_PATH = DIRS["outputs_metrics"] / f"bert_metrics_{RUN_TAG}.json"
BERT_PREDS_PATH = DIRS["outputs_metrics"] / f"bert_test_preds_{RUN_TAG}.npy"

if CFG["run_bert"]:
    if device != "cuda":
        raise RuntimeError("Treino do BERT requer GPU no Colab.")

    if not (BERT_BASE_FINAL_DIR / "config.json").exists():
        raise RuntimeError(
            f"Modelo final base não encontrado em {BERT_BASE_FINAL_DIR}. "
            "Primeiro confirme que o treino anterior foi salvo corretamente."
        )

    BERT_CONT_DIR.mkdir(parents=True, exist_ok=True)
    BERT_CONT_CKPT_DIR.mkdir(parents=True, exist_ok=True)

    # ---------------------------------------------------------
    # 1) Se esta continuação já foi concluída, apenas carrega
    # ---------------------------------------------------------
    if (BERT_CONT_DIR / "config.json").exists() and BERT_METRICS_PATH.exists():
        print("✅ Modelo BERT continuado já encontrado. Carregando...")
        bert_model = AutoModelForSequenceClassification.from_pretrained(str(BERT_CONT_DIR))
        metrics_bert = load_json(BERT_METRICS_PATH)

    else:
        # ---------------------------------------------------------
        # 2) Carrega do modelo final bom já salvo
        # ---------------------------------------------------------
        print(f"📦 Continuando a partir de: {BERT_BASE_FINAL_DIR}")

        bert_model = AutoModelForSequenceClassification.from_pretrained(
            str(BERT_BASE_FINAL_DIR),
            num_labels=3,
            id2label=ID2LABEL,
            label2id=LABEL2ID,
            local_files_only=True,
        )

        ta_params = inspect.signature(TrainingArguments.__init__).parameters

        args_dict = dict(
            output_dir=str(BERT_CONT_CKPT_DIR),
            num_train_epochs=CFG["bert_epochs"],   # aqui são épocas adicionais sobre o modelo final
            per_device_train_batch_size=CFG["bert_train_bs"],
            per_device_eval_batch_size=CFG["bert_eval_bs"],
            learning_rate=CFG["bert_lr"],
            weight_decay=0.01,
            save_strategy="steps",
            save_steps=1000,
            save_total_limit=3,
            logging_strategy="steps",
            logging_steps=50,
            load_best_model_at_end=True,
            metric_for_best_model="f1_macro",
            greater_is_better=True,
            fp16=(device == "cuda"),
            report_to="none",
            seed=SEED,
        )

        if "eval_strategy" in ta_params:
            args_dict["eval_strategy"] = "steps"
            args_dict["eval_steps"] = 1000
        else:
            args_dict["evaluation_strategy"] = "steps"
            args_dict["eval_steps"] = 1000

        args = TrainingArguments(**args_dict)

        trainer = Trainer(
            model=bert_model,
            args=args,
            train_dataset=hf_train,
            eval_dataset=hf_val,
            compute_metrics=compute_metrics,
        )

        print("🚀 Iniciando continuação limpa por mais épocas...")
        train_result = trainer.train()
        print("Loss final:", train_result.training_loss)

        # ---------------------------------------------------------
        # 3) Avaliação final no teste
        # ---------------------------------------------------------
        eval_test = trainer.evaluate(hf_test)

        # ---------------------------------------------------------
        # 4) Predições no teste
        # ---------------------------------------------------------
        preds_output = trainer.predict(hf_test)
        y_pred_bert_test = np.argmax(preds_output.predictions, axis=1)
        np.save(BERT_PREDS_PATH, y_pred_bert_test)
        print(f"💾 Predições salvas em: {BERT_PREDS_PATH}")

        # ---------------------------------------------------------
        # 5) Salvamento do novo modelo final
        # ---------------------------------------------------------
        print("💾 Salvando modelo final continuado...")
        trainer.save_model(str(BERT_CONT_DIR))
        tokenizer.save_pretrained(str(BERT_CONT_DIR))

        # ---------------------------------------------------------
        # 6) Histórico de avaliação
        # ---------------------------------------------------------
        eval_history = []
        for row in trainer.state.log_history:
            if "eval_f1_macro" in row:
                eval_history.append({
                    "step": int(row.get("step", 0)),
                    "epoch": float(row.get("epoch", 0)),
                    "f1": float(row["eval_f1_macro"]),
                    "acc": float(row.get("eval_accuracy", 0)),
                    "val_loss": float(row.get("eval_loss", 0)),
                })

        metrics_bert = {
            "model": "BERTimbau fine-tuned (continuação)",
            "base_model_dir": str(BERT_BASE_FINAL_DIR),
            "continued_model_dir": str(BERT_CONT_DIR),
            "run_tag": RUN_TAG,
            "additional_epochs": CFG["bert_epochs"],
            "f1_macro_test": float(eval_test["eval_f1_macro"]),
            "accuracy_test": float(eval_test["eval_accuracy"]),
            "eval_history": eval_history,
        }

        save_json(metrics_bert, BERT_METRICS_PATH)
        print("💾 Métricas da continuação salvas.")

    print(json.dumps(
        {
            "f1_macro_test": round(metrics_bert["f1_macro_test"], 4),
            "accuracy_test": round(metrics_bert["accuracy_test"], 4),
        },
        ensure_ascii=False,
        indent=2,
    ))
````

## 7. Bônus — In-Context Learning com Gemini

O experimento abaixo foi ajustado para:
- usar o **SDK atual** do Gemini;
- priorizar o modelo `gemini-3.1-flash-lite-preview`;
- respeitar **15 RPM**, com pausa explícita;
- gravar checkpoints de forma incremental;
- manter o custo e o tempo sob controle para entrega acadêmica.  

De acordo com a documentação oficial atual, o modelo `gemini-3.1-flash-lite-preview` é indicado para tarefas de classificação de alto volume e baixa latência. Ele está em preview e foi lançado em março de 2026.

A documentação oficial também mostra o uso do SDK Python atual via `from google import genai` e `client.models.generate_content(...)`.

Além disso, a série Gemini 3 suporta o parâmetro `thinking_level`, e o Flash-Lite aceita `minimal` e `low`, sendo apropriado reduzir o raciocínio para classificação simples a fim de baixar latência e custo.

````python
# 7.1 Configurar cliente Gemini (versão diagnosticável)

import os
from google.colab import userdata

gemini_client = None
GEMINI_API_KEY = None

def _mask_key(s: str) -> str:
    if not s:
        return "None"
    s = str(s)
    if len(s) <= 8:
        return "*" * len(s)
    return f"{s[:4]}...{s[-4:]}"

if CFG["run_gemini_icl"]:
    try:
        from google import genai

        # 1) Ler a secret explicitamente do Colab
        GEMINI_API_KEY = userdata.get(CFG["gemini_secret_name"])

        if not GEMINI_API_KEY:
            raise RuntimeError(
                f"Secret '{CFG['gemini_secret_name']}' não encontrada ou vazia no Colab."
            )

        # 2) Limpar possíveis variáveis concorrentes do runtime
        os.environ.pop("GOOGLE_API_KEY", None)
        os.environ.pop("GEMINI_API_KEY", None)

        # 3) Forçar a key correta no ambiente e no client
        os.environ["GEMINI_API_KEY"] = GEMINI_API_KEY
        gemini_client = genai.Client(api_key=GEMINI_API_KEY)

        print(f"✅ Cliente Gemini configurado com o modelo {CFG['gemini_model']}")
        print(f"🔑 Secret usada ({CFG['gemini_secret_name']}): {_mask_key(GEMINI_API_KEY)}")
        print("GOOGLE_API_KEY no ambiente:", "presente" if os.getenv("GOOGLE_API_KEY") else "ausente")
        print("GEMINI_API_KEY no ambiente:", "presente" if os.getenv("GEMINI_API_KEY") else "ausente")

    except Exception as e:
        print("⚠️ Não foi possível configurar o Gemini:", e)
        gemini_client = None
else:
    print("Execução Gemini desativada.")
````

````python
# 7.2 Prompting, parser e rate limiting

GEMINI_METRICS_PATH = DIRS["outputs_metrics"] / "gemini_icl_metrics.json"
GEMINI_PREDS_PATH = DIRS["outputs_metrics"] / "gemini_icl_preds.pkl"

FEW_SHOT_EXAMPLES = [
    ("Chegou bem embalado, funciona perfeitamente e atendeu todas as expectativas.", "positivo"),
    ("Produto excelente, entrega rápida e preço justo. Recomendo a todos.", "positivo"),
    ("Produto chegou com defeito e o suporte não resolveu. Péssima experiência.", "negativo"),
    ("Demorou 30 dias para chegar e veio errado. Não compro mais aqui.", "negativo"),
    ("O produto é mediano. Nada de especial, mas cumpre o que promete.", "neutro"),
]

SYSTEM_INSTRUCTION = (
    "Você classifica avaliações de produtos em português. "
    "Responda com apenas uma categoria: positivo, negativo ou neutro."
)

def build_prompt(texto: str, n_shots: int) -> str:
    shots = FEW_SHOT_EXAMPLES[:n_shots]
    parts = [SYSTEM_INSTRUCTION, ""]
    if shots:
        parts.append("Exemplos:")
        for ex_text, ex_label in shots:
            parts.append(f'Avaliação: "{ex_text}"')
            parts.append(f"Rótulo: {ex_label}")
            parts.append("")
    parts.append(f'Avaliação: "{texto}"')
    parts.append("Rótulo:")
    return "\n".join(parts)

def normalize_gemini_label(raw_text: str):
    if raw_text is None:
        return None
    s = str(raw_text).strip().lower()
    if "positivo" in s:
        return "positivo"
    if "negativo" in s:
        return "negativo"
    if "neutro" in s:
        return "neutro"
    return None

def predict_gemini_label(texto: str, n_shots: int):
    if gemini_client is None:
        return None

    prompt = build_prompt(texto, n_shots)
    last_err = None

    for attempt in range(CFG["gemini_max_retries"]):
        try:
            response = gemini_client.models.generate_content(
                model=CFG["gemini_model"],
                contents=prompt,
                config={
                    "temperature": 0,
                    "max_output_tokens": 8,
                    "thinking_config": {"thinking_level": "minimal"},
                },
            )
            raw = getattr(response, "text", None)
            label = normalize_gemini_label(raw)
            if label in LABEL2ID:
                return label
            return None
        except Exception as e:
            last_err = e
            wait = CFG["gemini_sleep_seconds"] * (attempt + 1)
            print(f"Retry {attempt+1}/{CFG['gemini_max_retries']} após {wait:.1f}s | erro: {e}")
            time.sleep(wait)

    print("Falha final Gemini:", last_err)
    return None

print("✅ Funções Gemini prontas.")
````

````python
# 7.3 Avaliação ICL no subconjunto estratificado do teste

if CFG["run_gemini_icl"]:
    n_eval = min(CFG["gemini_eval_size"], len(df_test))
    idx_eval, _ = train_test_split(
        np.arange(len(df_test)),
        train_size=n_eval,
        stratify=y_test,
        random_state=SEED,
    )

    df_gem = df_test.iloc[idx_eval].reset_index(drop=True)
    y_gem_true = y_test[idx_eval]

    if exists(GEMINI_METRICS_PATH) and exists(GEMINI_PREDS_PATH):
        metrics_gemini = load_json(GEMINI_METRICS_PATH)
        with open(GEMINI_PREDS_PATH, "rb") as f:
            gemini_preds_all = pickle.load(f)
        print("✅ Métricas ICL carregadas do cache.")
    else:
        metrics_gemini = {}
        gemini_preds_all = {}

        if gemini_client is None:
            print("⚠️ Cliente Gemini indisponível. Pulando avaliação ICL.")
        else:
            for n_shots in [0, 3, 5]:
                key = f"{n_shots}shot"
                preds = []
                print(f"\n▶ Avaliando {key} em {n_eval} exemplos...")

                for i, row in df_gem.iterrows():
                    pred_label = predict_gemini_label(row["text_clean"], n_shots=n_shots)
                    preds.append(pred_label)
                    if (i + 1) % 20 == 0:
                        valid = sum(p is not None for p in preds)
                        print(f"{i+1}/{n_eval} | válidas: {valid}")
                    time.sleep(CFG["gemini_sleep_seconds"])

                gemini_preds_all[key] = preds

                valid_mask = np.array([p in LABEL2ID for p in preds])
                if valid_mask.sum() == 0:
                    metrics_gemini[key] = {
                        "model": f"Gemini ICL {key}",
                        "n_shots": n_shots,
                        "n_eval": n_eval,
                        "n_valid": 0,
                        "f1_macro": None,
                        "accuracy": None,
                    }
                else:
                    y_true_valid = y_gem_true[valid_mask]
                    y_pred_valid = np.array([LABEL2ID[p] for p in np.array(preds, dtype=object)[valid_mask]])
                    metrics_gemini[key] = {
                        "model": f"Gemini ICL {key}",
                        "n_shots": n_shots,
                        "n_eval": n_eval,
                        "n_valid": int(valid_mask.sum()),
                        "f1_macro": float(f1_score(y_true_valid, y_pred_valid, average="macro")),
                        "accuracy": float(accuracy_score(y_true_valid, y_pred_valid)),
                    }

            save_json(metrics_gemini, GEMINI_METRICS_PATH)
            with open(GEMINI_PREDS_PATH, "wb") as f:
                pickle.dump(gemini_preds_all, f)
            print("💾 Resultados ICL salvos.")

    display(pd.DataFrame(metrics_gemini).T)
````

````python
# 8.2 Gráfico comparativo final

if len(df_results) > 0:
    df_plot = df_results.sort_values("F1-macro", ascending=True).copy()
    type_colors = {"Clássico": "#4e79a7", "Transformer": "#f28e2b", "LLM": "#59a14f"}

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    for ax, metric in zip(axes, ["F1-macro", "Acurácia"]):
        vals = df_plot[metric].values
        names = df_plot["Modelo"].values
        colors = [type_colors[t] for t in df_plot["Tipo"]]
        bars = ax.barh(names, vals, color=colors, alpha=0.9)
        ax.set_xlim(0, 1)
        ax.set_title(metric)
        for bar, val in zip(bars, vals):
            ax.text(val + 0.01, bar.get_y() + bar.get_height()/2, f"{val:.4f}", va="center")

    plt.tight_layout()
    save_fig(fig, "08_comparacao_final.png")
    plt.show()
````

````python
# 8.4.0 Recarregar artefatos corretos para C8.4 (hard cases)

import numpy as np
import joblib

# Paths corretos
TFIDF_MODEL_PATH = DIRS["models_tfidf"] / "tfidf_grid_best.pkl"
EMB_MODEL_PATH = DIRS["models_emb"] / "svm_embeddings.pkl"

TEST_EMB_PATH = DIRS["embeddings"] / "test_embeddings.npy"
BERT_PREDS_PATH = DIRS["outputs_metrics"] / "bert_test_preds_cont_2ep.npy"

# Rótulos reais do teste
y_test = df_test["label"].map(LABEL2ID).values

# TF-IDF
if TFIDF_MODEL_PATH.exists():
    tfidf_model = joblib.load(TFIDF_MODEL_PATH)
    y_pred_tfidf_test = tfidf_model.predict(df_test["text_clean"])
    print("✅ Predições TF-IDF recriadas.")
else:
    print(f"⚠️ Modelo TF-IDF não encontrado: {TFIDF_MODEL_PATH}")

# Embeddings + SVM
if EMB_MODEL_PATH.exists() and TEST_EMB_PATH.exists():
    emb_model = joblib.load(EMB_MODEL_PATH)
    X_test_emb = np.load(TEST_EMB_PATH)
    y_pred_emb_test = emb_model.predict(X_test_emb)
    print("✅ Predições embeddings recriadas.")
else:
    print(f"⚠️ Modelo embeddings não encontrado: {EMB_MODEL_PATH}")
    print(f"⚠️ Embeddings de teste não encontrados: {TEST_EMB_PATH}")

# BERT
if BERT_PREDS_PATH.exists():
    y_pred_bert_test = np.load(BERT_PREDS_PATH)
    print("✅ Predições BERT carregadas.")
else:
    print(f"⚠️ Predições BERT não encontradas: {BERT_PREDS_PATH}")

print(
    "Disponibilidade final:",
    "TF-IDF" if "y_pred_tfidf_test" in locals() else "TF-IDF ausente",
    "|",
    "EMB" if "y_pred_emb_test" in locals() else "EMB ausente",
    "|",
    "BERT" if "y_pred_bert_test" in locals() else "BERT ausente",
)
print("Tamanhos:",
      len(y_test),
      len(y_pred_tfidf_test) if "y_pred_tfidf_test" in locals() else None,
      len(y_pred_emb_test) if "y_pred_emb_test" in locals() else None,
      len(y_pred_bert_test) if "y_pred_bert_test" in locals() else None)
````
